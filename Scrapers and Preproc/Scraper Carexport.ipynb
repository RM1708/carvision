```python
import urllib2
from bs4 import BeautifulSoup
import re

base_url="https://www.carexportamerica.com/usa-cars-for-sale/?search_in=title&keyword=&minprice=&maxprice=&minyear=0&maxyear=0&stock=&today="

index_file=1
for page in range(0,20):
    r=urllib2.urlopen(base_url).read()
    soup=BeautifulSoup(r)
    links_cars=soup.find_all("img",height=150)
    for link in links_cars:
        src=link["src"]
        car_url=src.split("=",1)[1].split("&",1)[0]
        req=urllib2.Request(car_url)
        try:
            f=urllib2.urlopen(req)
        except urllib2.HTTPError, e:
            print(e.code)
        except urllib2.URLError, e:
            print(e.args)
        local_file=open("./pics_carexport/"+str(index_file)+".jpg","w")
        local_file.write(f.read()) #Write locally
        local_file.close
        index_file=index_file+1

    base_url=soup.find_all("a",class_="blue")[1]["href"]
    print(base_url)


    
    
#links_cars=soup.find_all("a",href=re.compile("on/cars/")) #links to cars in this page



```
