```python
#Pre-processing of images

from scipy import ndimage
from scipy import misc
import matplotlib.pyplot as plt
import os, sys
import numpy as np
import cv2
from __future__ import division

        

```

```python
#-------------------------CROP AND RESIZE IMAGES----------

#RESIZING PARAMETERS
target_ratio=0.70
target_width=400
target_height=int(400*0.625)

#Loop over all images
for filename in os.listdir("./pics_erep"):
    if filename.endswith("jpg"):
        pic=cv2.imread("./pics_erep/"+filename)
        print(filename)
        ratio=pic.shape[0]/(pic.shape[1])
        print(ratio)
        print(pic.shape)
        if (ratio>target_ratio):
            height_tar=pic.shape[1]*target_ratio
            print(height_tar)
            to_crop=int(round((pic.shape[0]-height_tar)/2))
            shift=int((pic.shape[0]-height_tar)%2)
            print(shift)
            print(to_crop)
            cropped=pic[to_crop-shift:-to_crop,]
        print(cropped.shape)

        if (ratio < target_ratio):
            width_tar=pic.shape[0]/target_ratio
            print(width_tar)
            to_crop=int(round((pic.shape[1]-width_tar)/2))
            shift=int((pic.shape[0]-height_tar)%2)
            print(width_tar)
            print(to_crop)
            #cropped=pic[,(to_crop):(-to_crop+shift)]
        print(cropped.shape)

        resized_pic = cv2.resize(cropped, (target_width,target_height))
        print(resized_pic.shape)
        cv2.imwrite("./pics_erepS/"+"DD"+filename,resized_pic)


```
